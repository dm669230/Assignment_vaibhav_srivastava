Q-5. A certain city has two taxi companies: Company A has 80% of the taxis and
Company B has 20% of the taxis. Company A's taxis have a 95% success rate for picking
up passengers on time, while Company B's taxis have a 90% success rate. If a randomly
selected taxi is late, what is the probability that it belongs to Company A?

Let's denote the following:
A: Event that the taxi belongs to Company A
B: Event that the taxi is late

We are looking for P(A|B), the probability that the taxi belongs to Company A given that it is late.

According to Bayes' theorem:
P(A|B) = (P(B|A) * P(A)) / P(B)

P(B|A) = Probability that the taxi is late given that it belongs to Company A = 1 - 0.95 = 0.05 (since 95% success rate means a 5% failure rate)
P(A) = Probability that the taxi belongs to Company A = 0.8
P(B) = Probability that the taxi is late

To calculate P(B), we need to consider both companies' probabilities of being late:
P(B) = P(B|A) * P(A) + P(B|B) * P(B)

P(B|B) = Probability that the taxi is late given that it belongs to Company B = 1 - 0.90 = 0.10 (since 90% success rate means a 10% failure rate)
P(B|B) = Probability that the taxi belongs to Company B = 0.2

Now we can substitute the values into Bayes' theorem:

P(A|B) = (0.05 * 0.8) / (0.05 * 0.8 + 0.10 * 0.2)

Simplifying the expression:

P(A|B) = 0.04 / (0.04 + 0.02)
= 0.04 / 0.06
= 2/3
≈ 0.67

Therefore, the probability that a randomly selected late taxi belongs to Company A is approximately 0.67 or 67%.

Q-10. A factory produces light bulbs, and the probability of a bulb being defective is 0.05.
The factory produces a large batch of 500 light bulbs.
a. What is the probability that exactly 20 bulbs are defective?
b. What is the probability that at least 10 bulbs are defective?
c. What is the probability that at max 15 bulbs are defective?
d. On average, how many defective bulbs would you expect in a batch of 500?

a. Probability of exactly 20 defective bulbs:


Using the binomial probability formula:

P(X = 20) = (500 C 20) * (0.05)^20 * (1 - 0.05)^(500 - 20)

Using the binomial coefficient formula (n C k) = n! / (k! * (n - k)!) to calculate (500 C 20):

P(X = 20) = (500! / (20! * (500 - 20)!)) * (0.05)^20 * (1 - 0.05)^(500 - 20)

Calculating this expression, we find that the probability of exactly 20 defective bulbs is approximately 0.0293.



b. Probability of at least 10 defective bulbs:


We can calculate the complement of the probability of having fewer than 10 defective bulbs:

P(X ≥ 10) = 1 - P(X < 10)

To calculate P(X < 10), we sum up the individual probabilities of having 0, 1, 2, ..., 9 defective bulbs using the binomial probability formula. Then we subtract this result from 1 to get P(X ≥ 10).

P(X < 10) = P(X = 0) + P(X = 1) + ... + P(X = 9)

Calculating this expression, we find the probability of having fewer than 10 defective bulbs is approximately 0.9867.

Therefore,
P(X ≥ 10) = 1 - 0.9867 = 0.0133

The probability of at least 10 defective bulbs is approximately 0.0133.



c. Probability of at most 15 defective bulbs:


We can calculate the cumulative probability up to 15 bulbs:

P(X ≤ 15) = P(X = 0) + P(X = 1) + ... + P(X = 15)

Calculating this expression, we find the probability of having at most 15 defective bulbs is approximately 0.9996.



d. Expected number of defective bulbs:


The average number of defective bulbs in a batch of 500 can be calculated using the mean of the binomial distribution:
    
Mean (μ) = n * p

Substituting the values:

μ = 500 * 0.05 = 25

The expected number of defective bulbs in a batch of 500 is 25.

Q-12. A pharmaceutical company develops a new drug and wants to compare its

study with two groups: Group A receives the new drug, and Group B receives the standard
drug. The company measures the improvement in a specific symptom for both groups after

Group A: Mean improvement score (x̄A) = 2.5, Standard deviation (sA) = 0.8, Sample size (nA) = 30.
Group B: Mean improvement score (x̄B) = 2.2, Standard deviation (sB) = 0.6, Sample size (nB) = 30.

We can calculate the t-statistic using the formula:
t = (x̄A - x̄B) / √[(sA² / nA) + (sB² / nB)]

Plugging in the values:
t = (2.5 - 2.2) / √[(0.8² / 30) + (0.6² / 30)]
= 0.3 / √[0.021333 + 0.012]

Next, we calculate the degrees of freedom:
df = [(sA² / nA + sB² / nB)²] / [(sA² / nA)² / (nA - 1) + (sB² / nB)² / (nB - 1)]
= [0.021333 + 0.012]² / [((0.021333)² / 29) + (0.012)² / 29]
≈ 655.75

Since we round down the degrees of freedom to the nearest integer, df = 655.

Using a significance level (α) of 0.05 and a two-tailed test, the critical t-value is approximately ±1.96 for df = 655.

Comparing the calculated t-statistic (0.3) with the critical t-value (±1.96), we find that the absolute value of the t-statistic is less than the critical t-value.

Therefore, we fail to reject the null hypothesis. In the context of the study, this means that there is no significant difference in the effectiveness of the new drug compared to the standard drug for treating the particular condition, based on the improvement in the specific symptom measured after the 4-week treatment period.
